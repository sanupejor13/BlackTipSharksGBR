```python
import pandas as pd
```

```python
with open('Old_records.csv', 'r') as in_file, open('cleanRecords.csv', 'w') as out_file:
    seen = set() # set for fast O(1) amortized lookup
    for line in in_file:
        if line in seen: continue # skip duplicate

       seen.add(line)
       out_file.write(line)
```

```python
sharks = pd.read_csv('cleanRecords.csv', sep = ';', on_bad_lines='skip', parse_dates=['eventDate'])
dfGBRlat = sharks[sharks['decimalLatitude'].between(-25, -5)]
dfGBR = dfGBRlat[dfGBRlat['decimalLongitude'].between(140, 160)].reset_index()
dfGBR.drop(['index', 'geodeticDatum'], axis=1).to_csv('GBRshark.csv')
```

```python
DF1 = pd.read_csv('GBRshark.csv', parse_dates=['eventDate'])
DF1
```

```python
Temp = pd.read_csv('temp_data.csv', sep = ';', parse_dates=['Aggregated Date/Time'] )
Temp
```

```python
Temp["day"] = Temp["Aggregated Date/Time"].dt.day
Temp["month"] = Temp["Aggregated Date/Time"].dt.month
Temp["year"] = Temp["Aggregated Date/Time"].dt.year
Temp
```

```python
dftemp = pd.merge(DF1,Temp, left_on=['year','month','day', 'decimalLatitude', 'decimalLongitude'],
                  right_on = ['year', 'month', 'day', 'Latitude', 'Longitude'], how='left')
dftemp
```

```python
T = dftemp.rename(columns={"mean":"Temperature"})
shark_temp = T.drop(columns=['eventDate', 'decimalLatitude', 'decimalLongitude'])
shark_temp
```

```python
Salt = pd.read_csv('salt_data.csv', sep = ';', parse_dates=['Aggregated Date/Time'] )
Salt
```

```python
Salt["day"] = Salt["Aggregated Date/Time"].dt.day
Salt["month"] = Salt["Aggregated Date/Time"].dt.month
Salt["year"] = Salt["Aggregated Date/Time"].dt.year
Salt
```

```python
dfsalt = pd.merge(shark_temp,Salt, left_on=['year','month','day','Latitude','Longitude'],
                  right_on = ['year','month','day','Latitude','Longitude'], how='left')
dfsalt
```

```python
S = dfsalt.rename(columns={"mean":"Salinity","Aggregated Date/Time_x":"Date"})
shark_data = S.drop(columns=['Aggregated Date/Time_y', 'Unnamed: 0','dataResourceUid'])
shark_data
```

```python
shark_data.to_csv('Shark_Temp_Salt.csv')
```
